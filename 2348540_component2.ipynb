```python
df = pd.DataFrame({'Country': ['India', 'USA', 'UK', 'Japan', 'China'],
                   'Happiness Score': [4.0, 7.5, 6.8, 5.5, 4.2]})

# Create bins
bins = [0, 4.9, 6.9, 10]

# Create labels for the bins
labels = ['Low Happiness', 'Medium Happiness', 'High Happiness']

# Add a new column with the bin labels
df['Happiness Group'] = pd.cut(df['Happiness Score'], bins=bins, labels=labels)

print(df)
```

```python


# create a sample dataframe
data = {'Region': ['North', 'South', 'East', 'West', 'North']}
df = pd.DataFrame(data)

# perform one-hot encoding
df_encoded = pd.get_dummies(df, columns=['Region'])

print(df_encoded)
```

```python
from sklearn.preprocessing import MinMaxScaler

# create a sample dataframe
data = {'Economy (GDP per Capita)': [1000, 765, 800]}
df = pd.DataFrame(data)

# perform Min-Max scaling
scaler = MinMaxScaler()
df['Economy (GDP per Capita)'] = scaler.fit_transform(df[['Economy (GDP per Capita)']])

print(df)
```

```python
from scipy.stats import chi2_contingency

# create a sample dataframe
data = {'Region': ['North', 'South', 'East', 'West', 'North'],
        'Happiness Rank': [1, 2, 3, 4, 5]}
df = pd.DataFrame(data)

# create a contingency table
contingency_table = pd.crosstab(df['Region'], df['Happiness Rank'])

# perform chi-square test
chi2_statistic, p_value, dof, expected_values = chi2_contingency(contingency_table)

print(f"Chi-square statistic: {chi2_statistic}")
print(f"P-value: {p_value}")
print(f"Degrees of freedom: {dof}")
print("Expected values:")
print(expected_values)
```
